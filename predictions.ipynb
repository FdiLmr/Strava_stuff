```python
import pandas as pd
```

## Reading data

```python
df = pd.read_csv('data/my_activity_data=20241115233935.csv')
print('Dataframe Shape:', df.shape)
df.head()
```

```python
# List all columns in the dataframe
print("Features in the dataset:")
for col in df.columns:
    print(f"- {col}")
```

### Features useful for performance prediction :
- id (useful to match the other csv)
- distance
- moving_time
- elapsed_time
- total_elevation_gain
- type
- average_heartrate
- max_heartrate
- suffer_score (=> relative effort)

```python
# Create a new DataFrame with selected features
selected_features = ['id', 'distance', 'moving_time', 'elapsed_time', 'total_elevation_gain', 
                    'type', 'average_heartrate', 'max_heartrate', 'suffer_score']

df_selected = df[selected_features].copy()

# Display the shape and first few rows of the new DataFrame
print('Selected Features DataFrame Shape:', df_selected.shape)
df_selected.head()
```

## Reading Activities data

```python
activities_df = pd.read_csv('data/activities.csv')
print('Activities Dataframe Shape:', activities_df.shape)
activities_df.head()
```

```python
# List all columns in the dataframe
print("Features in the dataset:")
for col in activities_df.columns:
    print(f"- {col}")
```

### Features useful for performance prediction :
- ID de l'activité
- Temps écoulé
- Distance
- Fréquence cardiaque maximum
- Fréquence cardiaque maximum.1
- Mesure d'effort
- Mesure d'effort.1
- Vitesse max.

- Dénivelé positif
- Dénivelé négatif

```python
# Create a new DataFrame with selected features from activities_df
selected_features_activities = [
    'ID de l\'activité',
    'Type d\'activité',
    'Temps écoulé',
    'Distance',
    'Fréquence cardiaque maximum',
    'Fréquence cardiaque maximum.1',
    'Mesure d\'effort',
    'Mesure d\'effort.1',
    'Vitesse max.',
    'Vitesse moyenne',
    'Dénivelé positif',
    'Dénivelé négatif',
    'Pente moyenne',
    'Fréquence cardiaque moyenne',
    'Charge d’entraînement',
    'Intensité'
]

activities_df_selected = activities_df[selected_features_activities].copy()

# Display the shape and first few rows of the new DataFrame
print('Selected Features Activities DataFrame Shape:', activities_df_selected.shape)
activities_df_selected.head()
```
